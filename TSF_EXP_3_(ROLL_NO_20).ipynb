**Import data**

```python
#DATA 1:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('/content/daily-total-female-births.csv')
data.plot(color="yellowgreen")
data.hist(color="yellowgreen")
plt.show()
```

```python
#split the data into three
split = round(len(data) / 3)
data1, data2, data3 = data[0:split], data[split:(2 *split)], data[(2 *split):]
mean1,mean2,mean3 = data1["Births"].mean(), data2["Births"].mean(),data3["Births"].mean()
var1, var2,var3 =  data1["Births"].var(), data2["Births"].var(),data3["Births"].var()
print("The means are :", str(round(mean1,2))+",",str(round(mean2,2))+",",round(mean3,2))
print("The variances are :", str(round(var1,2))+",",str(round(var2,2))+",",round(var3,2))
```
